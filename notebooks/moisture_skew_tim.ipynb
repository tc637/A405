```python
from importlib import reload
import a405thermo.thermlib
reload(a405thermo.thermlib)
from a405thermo.thermlib import find_esat

def zero_rs(temp,rsat,press):
    """
      find the saturation temperature for 
      a given rsat,press, by rootfinding this zero
      
      input: temp (guess) (K)
             rsat (kg/kg)
             press (hPa)
      output: residual
      
      see thompkins 2.20
     
    """
    esat=find_esat(temp)*0.01  #convert to hPa
    residual=rsat - c.eps*esat/(press - esat)
    return residual

def find_rsat(temp,press):
    """
       input: temp (K)
              press (hPa)
        output: rsat (kg/kg)
    """
    esat = find_esat(temp)*0.01
    rsat=c.eps*esat/(press - esat)
    return rsat

def zero_find_rs(tstart,rsat,press):
    """
       find the temp that produces rsat at press.
       input: temp (K)
              rsat (kg/kg)
              press (hPa)
        output: temp (K)
    """
    brackets=rf.find_interval(zero_rs,tstart,rsat,press)
    temp = rf.fzero(zero_rs,brackets,rsat,press)
    return temp
```

```python
from a405thermo.thermlib import find_lcl,find_rsat,find_thetaes,find_Tmoist,find_rvrl
def calc_buoy(ind, temp, pres, rt, thetae):
    """Calculates integrand of the equation for CAPE at a given pressure level
    
        Inputs:
            ind (int) - Height index
            temp (float) - Temperature (K) at height index ind
            pres (float) - Pressure (Pa) at height index ind
            rt (float) - Total mixing ratio (kg kg^-1)
            thetaes (float) - Pseudoadiabat (K)
            
        Outputs:
            buoy (float) - Buoyancy (m/s^2)
    
    """
    the_temp = temp[ind] # K
    the_pres = pres[ind] # Pa 
    grav = 9.81 # m/s^2
    
    moist_temp = find_Tmoist(thetae,the_pres) # calculate temperature at height index for pseudoadiabat
    
    # calculate Tv of the sounding and of the environment
    rv,rl = find_rvrl(the_temp, rt, the_pres)

    tv_env = (the_temp)*(1+0.608*rv-rl)
    tv_thetae = (moist_temp)*(1+0.608*rv-rl)
    
    # calculate "buoyancy" at that height
    buoy = grav*((tv_thetae - tv_env)/tv_env)

    return buoy
```

```python
sfc_pres = sounding["pres"][0]*100. # Pa
sfc_temp = sounding["temp"][0]+273.15 # K
sfc_dwpt = sounding["dwpt"][0]+273.15 # K

# find LCL and pseudoadiabat
[temp_lcl, press_lcl] = find_lcl(sfc_dwpt, sfc_temp, sfc_pres) # [(K), (Pa)]
thetae = find_thetaes(temp_lcl, press_lcl) # K

# plot LCL location to confirm
skew_lcl=convertTempToSkew(temp_lcl-c.Tc, press_lcl/100., skew)
ax.plot(skew_lcl,press_lcl/100,"om",label='lcl',markersize=12)
display(fig)

# find presssures and temperatures above LCL
inds = np.where(sounding['pres']*100<press_lcl)[0] # pressure indices above the LCL
sat_pres = sounding['pres'][inds]*100
sat_temp = sounding['temp'][inds]+c.Tc # K

# find rt
rt = find_rsat(sfc_dwpt, sfc_pres) # kg kg^-1

# append integrand to buoy array
buoy = []
for ind in inds:
    buoy.append(calc_buoy(ind, sat_temp, sat_pres, rt, thetae))
```

```python
%matplotlib inline
plt.style.use("ggplot")

figb, axb = plt.subplots(1,1,figsize=(6,6))
axb.plot(buoy, sat_pres/100)
axb.invert_yaxis() # pressure decreases with height
axb.set_ylim([1000,200]) # limit to plotting troposphere
axb.set_xlim([-0.15,0.15])

axb.set_title("Plot of Buoyancy (CAPE Integrand) as a Function of Height")
axb.set_ylabel("P (hPa)")
axb.set_xlabel(r"B ($m/s^2$)")
```
