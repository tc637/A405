```python
with pd.HDFStore(outfile,'r') as store:
    sounding=store['Y2015_Sep_12_00Z']
    print(sounding.loc[:5])
```

Lohmann 3.7:

$$ \frac {dP} {P} = -\frac {g} {R_d\overline{T_V}} dz$$ 

$$ \int_{P_0}^{P} \frac {dP} {P} = -\int_{z_0}^{z} \frac {g} {R_d\overline{T_V}} dz $$

$$ H = \frac {R_d\overline{T_V}} {g} = \frac {-\int_{z_0}^{z} dz} {\int_{P_0}^{P} \frac {dP} {P}} $$

```python
pressures = sounding['pres']*100. # convert to Pa from hPa
heights = sounding["hght"] # m

int_pressures = np.sum(np.diff(pressures)/pressures[0:-1]) # len(pressures) = len(np.diff(pressures)) + 1
int_heights = np.sum(np.diff(heights))

scale_height = -int_heights/int_pressures # m

print("The scale height is {:.3f} km for this sounding.".format(scale_height/1000))
```

Isothermal atmosphere:

$$ P = P_0\large e^{-\frac {z} {H}} $$

```python
P0 = 1000. # hPa
isotherm_pressures = P0*np.exp(-heights/scale_height)
```

```python
%matplotlib inline
m2km=1.e-3  #convert meters to km
plt.style.use('ggplot')
fig,ax=plt.subplots(1,1,figsize=(8,10))
ax.plot(sounding['pres'],sounding['hght']*m2km,label='Sounding')
ax.plot(isotherm_pressures,sounding['hght']*m2km,label='Isothermal Atmosphere')
out=ax.set(xlabel="pressure (hPa)",ylabel="height (km)",
      title ="September 2015 Pressure Curves, Hilo International")
ax.legend()
```
