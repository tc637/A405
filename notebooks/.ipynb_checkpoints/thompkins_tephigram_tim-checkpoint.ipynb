Thompkins (1.46):

$$ ln\theta = lnT + \frac {R_d} {c_p} (lnp_0 - lnp) $$


```python
import numpy as np
import matplotlib.pyplot as plt

def calc_ln_theta(temp, pres):
    """
    
    Inputs: temp (K), array of temperatures
            pres (hPa), pressure value, float
            
    Outputs: ln(theta), array of logged potential temperatures"""
    
    p0 = 100000. # Pa
    pres = pres*1000. # convert pres to Pa
    
    Rd = 287. # J kg^-1 K^-1
    cp = 1004. # J kg^-1 K^-1
    
    return(np.log(temp) + Rd/cp*(np.log(p0) - np.log(pres)))
```

```python
%matplotlib inline
dT = 0.5 # temperature interval, K
temps = np.arange(250, 300 + dT, dT) # K
pres = [1000, 800, 600] # hPa

fig, ax = plt.subplots(1,1, figsize=(6,6))

for each_pres in pres: # loop over pressures
    log_theta = calc_ln_theta(temps, each_pres)
    ax.plot(temps, log_theta, label=(str(each_pres) + " hPa")) # plot onto axis

ax.set_title("Replication of Thompkins Fig. 1.9 Tephigram (Isobars)")
ax.set_ylabel(r"$ln\theta$", fontsize=20)
ax.set_xlabel("T (K)", fontsize=20)
ax.legend(loc = "best")
ax.grid(b=True, which='major', color='k', linestyle='-')
plt.show()
```
